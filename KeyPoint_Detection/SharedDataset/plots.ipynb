```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/420px/training_output/results.csv', delimiter=',')
epochs = df['epoch'].values
box_loss = df['train/box_loss'].values
pose_loss = df['train/pose_loss'].values
kobj_loss = df['train/kobj_loss'].values
cls_loss = df['train/cls_loss'].values
dfl_loss = df['train/dfl_loss'].values

loss = 3.0 * box_loss + 0.5 * cls_loss + 0.6 * dfl_loss + 25.0 * pose_loss + 6.0 * kobj_loss

plt.figure(figsize=(10, 5))
plt.plot(epochs, loss, label='Total Training Loss', color='red', marker='')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.title(r'Training Loss Curve - $416 \, \mathrm{px}$')
plt.legend()
plt.grid()
#plt.savefig('training_loss_curve.png')
plt.show()
```

```python
box_loss_val = df['val/box_loss'].values
pose_loss_val = df['val/pose_loss'].values
kobj_loss_val = df['val/kobj_loss'].values
cls_loss_val = df['val/cls_loss'].values
dfl_loss_val = df['val/dfl_loss'].values

loss_val = 3.0 * box_loss_val + 0.5 * cls_loss_val + 0.6 * dfl_loss_val + 25.0 * pose_loss_val + 6.0 * kobj_loss_val

plt.figure(figsize=(10, 5))
plt.plot(epochs, loss_val, label='Total Validation Loss', color='blue', marker='')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Curve - $416 \, \mathrm{px}$')
plt.legend()
plt.grid()
#plt.savefig('validation_loss_curve.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# leggi i file salvati
f1_matrix416 = pd.read_csv(
    '/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/420px/F1_matrix.csv',
    header=None
).values

thresholds416 = np.loadtxt('/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/420px/F1_axis_thresholds_px.csv', delimiter=',')
confidences416 = np.loadtxt('/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/420px/F1_axis_confidences.csv', delimiter=',')

f1_matrix800 = pd.read_csv(
    '/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/800px/F1_matrix.csv',
    header=None
).values

thresholds800 = np.loadtxt('/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/800px/F1_axis_thresholds_px.csv', delimiter=',')
confidences800 = np.loadtxt('/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/800px/F1_axis_confidences.csv', delimiter=',')


def plot_F1_surfaces(
    thresholds416, confidences416, f1_matrix416,
    thresholds800, confidences800, f1_matrix800,
    save_img=False, filename="F1_surface_comparison.png"
):
    fig = plt.figure(figsize=(20, 7))

    # Primo plot (416)
    ax1 = fig.add_subplot(1, 2, 1, projection='3d')
    X1, Y1 = np.meshgrid(thresholds416, confidences416)
    surf1 = ax1.plot_surface(X1, Y1, f1_matrix416, cmap='viridis', edgecolor='k', alpha=0.8)
    ax1.set_xlabel('Threshold (px)')
    ax1.set_ylabel('Confidence Threshold')
    ax1.set_zlabel('F1')
    ax1.set_title('F1 Surface - 416px')
    ax1.view_init(elev=30, azim=50)
    fig.colorbar(surf1, ax=ax1, shrink=0.6, aspect=10)

    # Secondo plot (800)
    ax2 = fig.add_subplot(1, 2, 2, projection='3d')
    X2, Y2 = np.meshgrid(thresholds800, confidences800)
    surf2 = ax2.plot_surface(X2, Y2, f1_matrix800, cmap='viridis', edgecolor='k', alpha=0.8)
    ax2.set_xlabel('Threshold (px)')
    ax2.set_ylabel('Confidence Threshold')
    ax2.set_zlabel('F1')
    ax2.set_title('F1 Surface - 800px')
    ax2.view_init(elev=30, azim=50)
    fig.colorbar(surf2, ax=ax2, shrink=0.6, aspect=10)

    plt.tight_layout()

    if save_img:
        plt.savefig(filename, dpi=300)

    plt.show()


plot_F1_surfaces(
    thresholds416, confidences416, f1_matrix416,
    thresholds800, confidences800, f1_matrix800,
    save_img=True, filename="F1_surface_comparison.png"
)
```

```python
def plot_F1_vs_confidence(thresholds416, confidences416, f1_matrix416,
                          thresholds800, confidences800, f1_matrix800,
                          target_threshold416=4, target_threshold800=8):
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

    # --- 416px ---
    idx416 = np.where(thresholds416 == target_threshold416)[0][0]
    f1_at_threshold416 = f1_matrix416[:, idx416]
    axes[0].plot(confidences416, f1_at_threshold416, marker='.', color='green')
    axes[0].set_xlabel("Confidence Threshold")
    axes[0].set_ylabel("F1 score")
    axes[0].set_title(f"F1 and Confidence Threshold at Fixed Threshold ({target_threshold416} px) - 416px")
    axes[0].grid(True)
    max_conf416 = confidences416[np.argmax(f1_at_threshold416)]
    print(f'[416px] Max F1 = {np.max(f1_at_threshold416):.3f} at confidence = {max_conf416:.2f}')

    # --- 800px ---
    idx800 = np.where(thresholds800 == target_threshold800)[0][0]
    f1_at_threshold800 = f1_matrix800[:, idx800]
    axes[1].plot(confidences800, f1_at_threshold800, marker='.', color='blue')
    axes[1].set_xlabel("Confidence Threshold")
    axes[1].set_title(f"F1 and Confidence Threshold at Fixed Threshold ({target_threshold800} px) - 800px")
    axes[1].grid(True)
    max_conf800 = confidences800[np.argmax(f1_at_threshold800)]
    print(f'[800px] Max F1 = {np.max(f1_at_threshold800):.3f} at confidence = {max_conf800:.2f}')

    plt.tight_layout()
    
    plt.savefig('F1_vs_confidence_comparison.png')
    
    plt.show()


plot_F1_vs_confidence(
    thresholds416, confidences416, f1_matrix416,
    thresholds800, confidences800, f1_matrix800
)
```

```python
def plot_F1_vs_confidence_multi(thresholds416, confidences416, f1_matrix416,
                                thresholds800, confidences800, f1_matrix800):
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 9), sharey=True)

    # --- 416px ---
    for i, thr in enumerate(thresholds416):
        axes[0].plot(confidences416, f1_matrix416[:, i], marker='.', label=f'{thr:.1f} px')
    axes[0].set_xlabel("Confidence Threshold")
    axes[0].set_ylabel("F1 score")
    axes[0].set_title("F1 vs Confidence (various thresholds) - 416px")
    axes[0].legend(title="Threshold", fontsize=9)
    axes[0].grid(True)

    # --- 800px ---
    for i, thr in enumerate(thresholds800):
        axes[1].plot(confidences800, f1_matrix800[:, i], marker='.', label=f'{thr:.1f} px')
    axes[1].set_xlabel("Confidence Threshold")
    axes[1].set_title("F1 vs Confidence (various thresholds) - 800px")
    axes[1].legend(title="Threshold", fontsize=9)
    axes[1].grid(True)

    plt.tight_layout()
    
    plt.savefig('F1_vs_confidence_multi_comparison.png')
    
    plt.show()


plot_F1_vs_confidence_multi(
    thresholds416, confidences416, f1_matrix416,
    thresholds800, confidences800, f1_matrix800
)
```

```python
def plot_avg_keypoints(confidences416, average_pred416, f1_at_thr416,
                       confidences800, average_pred800, f1_at_thr800):

    fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

    # --- 416px ---
    axes[0].plot(confidences416, average_pred416, marker='.', color='red',
                 label='Average Predicted Keypoints')
    axes[0].axhline(y=160, color='blue', linestyle='--',
                    label='Mean Average Ground Truth Keypoints')
    axes[0].axhline(y=average_pred416[np.argmax(f1_at_thr416)], color='lime',
                    linestyle='--', label='Average Predicted Keypoints at Optimal Confidence')
    axes[0].set_xlabel("Confidence Threshold")
    axes[0].set_ylabel("Average Number of Predicted Keypoints")
    axes[0].set_title("Average Predicted Keypoints vs Confidence - 416px")
    axes[0].legend()
    axes[0].grid(True)

    # --- 800px ---
    axes[1].plot(confidences800, average_pred800, marker='.', color='red',
                 label='Average Predicted Keypoints')
    axes[1].axhline(y=160, color='blue', linestyle='--',
                    label='Mean Average Ground Truth Keypoints')
    axes[1].axhline(y=average_pred800[np.argmax(f1_at_thr800)], color='lime',
                    linestyle='--', label='Average Predicted Keypoints at Optimal Confidence')
    axes[1].set_xlabel("Confidence Threshold")
    axes[1].set_title("Average Predicted Keypoints vs Confidence - 800px")
    axes[1].legend()
    axes[1].grid(True)

    plt.tight_layout()
    
    plt.savefig('Avg_pred_keypoints_vs_confidence_comparison.png')
    
    plt.show()

    print(f'416px: Average predicted keypoints at optimal confidence = {average_pred416[np.argmax(f1_at_thr416)]:.2f}')
    print(f'800px: Average predicted keypoints at optimal confidence = {average_pred800[np.argmax(f1_at_thr800)]:.2f}')



average_pred416 = pd.read_csv(
    '/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/420px/avg_preds_per_conf.csv',
    header=None
).values.flatten()

average_pred800 = pd.read_csv(
    '/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/800px/avg_preds_per_conf.csv',
    header=None
).values.flatten()


plot_avg_keypoints(confidences416, average_pred416, f1_matrix416[:, np.where(thresholds416 == 4)[0][0]],
                   confidences800, average_pred800, f1_matrix800[:, np.where(thresholds800 == 8)[0][0]])
```
