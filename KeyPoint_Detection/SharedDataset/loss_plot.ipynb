```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/user/gr1/delphi/dghezzi/KeyPoint_Detection/SharedDataset/420px/training_output/results.csv', delimiter=',')
epochs = df['epoch'].values
box_loss = df['train/box_loss'].values
pose_loss = df['train/pose_loss'].values
kobj_loss = df['train/kobj_loss'].values
cls_loss = df['train/cls_loss'].values
dfl_loss = df['train/dfl_loss'].values

loss = 3.0 * box_loss + 0.5 * cls_loss + 0.6 * dfl_loss + 25.0 * pose_loss + 6.0 * kobj_loss

plt.figure(figsize=(10, 5))
plt.plot(epochs, loss, label='Total Training Loss', color='red', marker='')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.title(r'Training Loss Curve - $416 \, \mathrm{px}$')
plt.legend()
plt.grid()
#plt.savefig('training_loss_curve.png')
plt.show()
```

```python
box_loss_val = df['val/box_loss'].values
pose_loss_val = df['val/pose_loss'].values
kobj_loss_val = df['val/kobj_loss'].values
cls_loss_val = df['val/cls_loss'].values
dfl_loss_val = df['val/dfl_loss'].values

loss_val = 3.0 * box_loss_val + 0.5 * cls_loss_val + 0.6 * dfl_loss_val + 25.0 * pose_loss_val + 6.0 * kobj_loss_val

plt.figure(figsize=(10, 5))
plt.plot(epochs, loss_val, label='Total Validation Loss', color='blue', marker='')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Curve - $416 \, \mathrm{px}$')
plt.legend()
plt.grid()
#plt.savefig('validation_loss_curve.png')
plt.show()
```
