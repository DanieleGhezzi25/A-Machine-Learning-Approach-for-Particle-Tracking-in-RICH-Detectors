```python
import sys

sys.path.append(r'/user/gr1/delphi/dghezzi/KeyPoint_Detection')

import KP as kpl
import numpy as np

image_path = r'/user/gr1/delphi/dghezzi/KeyPoint_Detection/Models/Prova.jpg'
txt_path = r'/user/gr1/delphi/dghezzi/KeyPoint_Detection/Models/Prova.txt'
model_path = 'last.pt'

results = kpl.inference(image_path, model_path, confidence=0.5, show_image=True, save_image=True)
kpl.show_with_MCpoints(results, image_path, txt_path, show_image=True, save_image=True)
```

```python
images_dir = r'/user/gr1/delphi/dghezzi/KeyPoint_Detection/Models/TestDataset/images/train'
labels_dir = r'/user/gr1/delphi/dghezzi/KeyPoint_Detection/Models/TestDataset/labels/train'

kpl.inference_setImages(images_dir, labels_dir, model_path, confidence=0.3, save=True)
```

```python
pck_interval = np.arange(1, 10, 1)  # PCK thresholds in pixels
conf_interval = np.arange(0.2, 0.9, 0.1)  # Confidence thresholds

F1_matrix = kpl.inference_3Dmap(images_dir, labels_dir, model_path,
                                 pck_thresholds=pck_interval,
                                 conf_thresholds=conf_interval)

print("F1 Matrix:")
print(F1_matrix)
```

```python
import pandas as pd

df = pd.DataFrame(F1_matrix, index=conf_interval, columns=pck_interval)
df.index.name = 'Confidence Threshold'
df.columns.name = 'PCK Threshold (px)'

df.to_csv('F1_matrix_labeled_lastpt.csv')
```
